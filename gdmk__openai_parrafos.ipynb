```python
import openai
from docx import Document
import pandas as pd
import os
```

```python
f = open('OAI_2023_09F.txt','r')
ftx = f.readlines()
tok = ''.join(ftx).replace('\n', '')
openai.api_key = tok
```

```python
ruta = 'drive/My Drive/2023 Cursos/Capacidades Cursos/cursos_LLM_codigos/'
print(os.listdir(ruta))
```

```python
archivo = 'parrafos_quijote.xlsx'
```

```python
base_textos = pd.read_excel(ruta + archivo)
```

```python
base_textos
```

```python
prompt_sistema = "Eres una experta en identificación de actividades en textos literarios."

terminos = """
Los términos o definiciones a tomar en cuenta para el análisis son:

Actividades Mentales: Cuando se alude a algun ejercicio mental de los protagonistas del texto, como razonamientos, pensamientos, sueños, ideas u otros
Actividades Físicas: Cuando se alude a algún movimiento o ejercicio de fuerza física de los protagonistas

"""

pedido = """
Tomando en cuenta el texto a analizar, detectar y reportar dos listas: en una lista separada por comas cuáles son las actividades mentales físicas detectadas, y en otra lista separada por comas cuáles son las actividades mentales detectadas
"""
```

```python
respuestas = []
for j, row in base_textos.iterrows():
    print(row.item, end=' ')
    tx_x = row.fragmento

    prompt_x = terminos + pedido + f"""El texto a analizar es: \"\"\"{tx_x}\"\"\" """

    response = openai.ChatCompletion.create(
      model="gpt-3.5-turbo-16k",
      messages=[
        {"role": "system", "content": prompt_sistema},
        {"role": "user", "content": prompt_x},

      ],
      temperature=0.5,
      max_tokens=300,
      top_p=1
    )


    respuesta = response['choices'][0].text.replace('\n','')
    print(response)
    respuestas.append([row.item, respuesta])
resp_df = pd.DataFrame(respuestas)
```
